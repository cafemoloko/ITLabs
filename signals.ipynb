```python
import matplotlib.pyplot as plot 
import numpy as np 
```

## Example 1

```python
freq = 3             # frequency
fs = 150             # sampling rate
phi = 3 * np.pi      # phase
n = np.arange(500)   # number of samples
```

```python
sf = 0.9 * np.sin(2 * np.pi * freq * (n / fs) * np.cos(np.pi * (n / fs) + phi))
```

```python
plot.plot(n, sf)
plot.xlabel(' no of sample')
plot.ylabel('phase angle')
plot.title('sinusoidal course')
plot.show()
```

## Example 2

```python
y = np.exp(-n * 0.04)
z = 0.001 * sf + 0.01 * y
```

```python
plot.plot(n, z)
plot.xlabel(' no of sample')
plot.ylabel('angle')
plot.title('[y = e^(-n/25)] signal course')
plot.show()
```

```python
v = sf * y
```

```python
plot.plot(n, v)
plot.xlabel(' no of sample')
plot.ylabel('angle')
plot.title('[y = e^(-n/25)] signal course')
plot.show()
```

## Example 3

```python
fs = 2300     # sampling rate
T = 1.8       # signal length
n = fs * T
```

```python
t = np.arange(1, n+1) / fs
index = (np.array([0.2, 0.7]) * fs).astype(int)   # bins
```

```python
t1 = t[0:index[0]]
t2 = t[index[0]:index[1]]
t3 = t[index[1]:]
```

```python
a1 = 50 * np.pi * t1
a1 = (t1 * 0.5) * np.sin(a1)
```

```python
a2 = 34 * np.pi * t2
a2 = (1 / t2) * 0.8 * np.sin(a2)
```

```python
a3 = 20 * np.pi * t3
a3 = 0.2 * np.sin(2 * a3) * np.sin(a3)
```

```python
a = np.concatenate((a1, a2, a3))
```

```python
plot.plot(np.arange(n), a)
plot.xlabel('no of sample')
plot.ylabel('angle')
plot.title('signal course')
plot.show()
```

## Example 4

```python
T = 4               # signal length
fs = 10000          # sampling rate
H = [3, 6, 60]
```

```python
for h in H:
    n = np.arange(1, h + 1) 
    t = np.arange(1, (fs * T) + 1) / fs 
    s = np.tensordot(n, t, axes=0) * np.pi
    m = np.sin(s / 2) * np.sin(20 * s)
    res = (9/np.pi ** 2) * np.sum(m, axis=0)
```

```python
plot.plot(np.arange(len(res)), res)
plot.xlabel('no of sample')
plot.ylabel('angle')
plot.title('signal course')
plot.show()
```
